```python
import pandas as pd
```

```python
dt1 = pd.read_csv('API_SH.XPD.TOTL.ZS_DS2_en_csv_v2.csv',skiprows = 3)
dt1.head()
```

```python
# find columns without any data
null_cols = dt1.apply(lambda x: sum(x.isnull()), axis=0)
null_cols = null_cols[null_cols==dt1.shape[0]]

# columns with at least one value
value_cols = []
total_cols = dt1.columns
for col in total_cols:
    if col not in null_cols.index:
        value_cols.append(col)
print(value_cols)
```

```python
final_cols = [col for col in value_cols if col not in ['Indicator Name', 'Indicator Code']]

df1 = dt1[final_cols]
df1.head()
```

```python
dt2 = pd.read_csv('Metadata_Country_API_SH.XPD.TOTL.ZS_DS2_en_csv_v2.csv')
dt2.head()
```

```python
df2 = dt2[['Country Code','Region','IncomeGroup'] ]
df2.head()
```

```python
df = pd.merge(df2,df1, how='inner', on=['Country Code', 'Country Code'])
df.head()
```
