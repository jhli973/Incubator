```python
import pandas as pd
import numpy as np
```

```python
from plotly.offline import init_notebook_mode, iplot, plot
from plotly.graph_objs import Scatter, Figure, Layout
init_notebook_mode(connected=True)
```

```python
df_in= pd.read_csv('HealthExpenditurebyIncomeGroup.csv')
df_in.head()
```

```python
dt_hi = Scatter(x=df_in.Year, y=df_in['High income'], name="High Income", line=dict(color='#0A0AGG'))
dt_li = Scatter(x=df_in.Year, y=df_in['Low income'] , name="Low Income" )
dt_umi = Scatter(x=df_in.Year, y=df_in['Upper middle income'], name="Upper Middle Income")
dt_lmi = Scatter(x=df_in.Year, y=df_in['Lower middle income'], name="Lower Middle Income" )

data = [dt_hi, dt_li, dt_umi, dt_lmi]
layout = dict(title="Health Expenditure by Income", titlefont=dict(size =14, color='black', family='Arial, sans-serif'),
              yaxis=dict(title='Expenditure (% GDP)', showline=True,ticklen=8,tick0=4, dtick=1,linewidth=2),
              xaxis=dict(title='Year', tick0=1995, dtick=2, tickangle=45, showline=True,ticklen=8,linewidth=2)
              #,legend=dict(x=-0.1, y=1)
             )
fig = dict(data=data, layout=layout)
iplot(fig, filename='hebyic_plotly.png')
```

```python
dt_eap = Scatter(x=df_re.Year, y=df_re['East Asia & Pacific'], name="East Asia & Pacific")
dt_eca = Scatter(x=df_re.Year, y=df_re['Europe & Central Asia'] , name="Europe & Central Asia" )
dt_lac = Scatter(x=df_re.Year, y=df_re['Latin America & Caribbean'], name="Latin America & Caribbean")
dt_mena = Scatter(x=df_re.Year, y=df_re['Middle East & North Africa'], name="Middle East & North Africa" )
dt_na = Scatter(x=df_re.Year, y=df_re['North America'] , name="North America" )
dt_sa = Scatter(x=df_re.Year, y=df_re['South Asia'], name="South Asia")
dt_ssa = Scatter(x=df_re.Year, y=df_re['Sub-Saharan Africa'], name="Sub-Saharan Africa" )

data = [dt_eap, dt_eca, dt_lac, dt_mena, dt_na, dt_sa, dt_ssa]
layout = dict(title="Health Expenditure by Region", titlefont=dict(size =14, color='black', family='Arial, sans-serif'),
              yaxis=dict(title='Expenditure (% GDP)', showline=True,ticklen=8,tick0=4, dtick=2,linewidth=2),
              xaxis=dict(title='Year', tick0=1995, dtick=2, tickangle=45, showline=True,ticklen=8,linewidth=2)
              #,legend=dict(x=-0.1, y=1)
             )
fig = dict(data=data, layout=layout)
iplot(fig, filename='hebyrg_plotly.png')
```
