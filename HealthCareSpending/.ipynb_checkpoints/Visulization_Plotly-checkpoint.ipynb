```python
import pandas as pd
import numpy as np
```

```python
from plotly.offline import init_notebook_mode, iplot, plot
from plotly.graph_objs import Scatter, Figure, Layout
init_notebook_mode(connected=True)
```

```python
df= pd.read_csv('HealthExpenditurebyIncomeGroup.csv')
df.head()
```

```python
dt_hi = Scatter(x=df.Year, y=df['High income'], name="High Income", line=dict(color='#0A0AGG'))
dt_li = Scatter(x=df.Year, y=df['Low income'] , name="Low Income" )
dt_umi = Scatter(x=df.Year, y=df['Upper middle income'], name="Upper Middle Income")
dt_lmi = Scatter(x=df.Year, y=df['Lower middle income'], name="Lower Middle Income" )

data = [dt_hi, dt_li, dt_umi, dt_lmi]
layout = dict(title="Health Expenditure by Income",
              yaxis=dict(title='Expenditure (% GDP)'),
              xaxis=dict(title='Year')
              #,legend=dict(x=-0.1, y=1)
             )
fig = dict(data=data, layout=layout)
iplot(fig, filename='hebyic_plotly.png')
```
